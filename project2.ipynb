```python
import matplotlib.pyplot as plt
t=sns.load_dataset('penguins')
t.head(100)
```

```python
plt.figure(figsize=(10,4))
sns.countplot(x="species",data=t,color='green')
plt.title('Penguins count by class')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x="species",y='body_mass_g',data=t,color='green')
plt.title('Body mass by Species')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(t['body_mass_g'],bins=20,kde=True,color='red')
plt.title('Distribution of Body Mass')
plt.show()
```

```python
counts=t['sex'].value_counts()
counts
```

```python
gender_counts=t['sex'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(gender_counts,labels=['Male','Female'],colors=['red','green'],autopct='%1.1f%%')
plt.title('Gender distribution of penguins')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='body_mass_g',data=t)
plt.title('Body mass by Species')
plt.xlabel('')
plt.ylabel('Body Mass')
plt.show()
```

```python
sns.scatterplot(x='bill_length_mm',y='flipper_length_mm',data=t)
plt.title('Bill length vs Flipper length')
plt.xlabel('Flipper length')
plt.ylabel('Bill length')
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x='bill_length_mm',y='body_mass_g',hue='species',style='sex',data=t)
plt.title('Bill length vs Body mass with Speicies and Gender')
plt.xlabel('Body mass')
plt.ylabel('Bill length')
plt.show()
```

```python
numerical=t[['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']].corr()
plt.figure(figsize=(4,4))
sns.heatmap(numerical,annot=True,cmap='rainbow')
plt.title('Correlation Heatmap of Penguin Dataset')
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.violinplot(x='species',y='body_mass_g',hue='sex',data=t,split=True)
plt.title('Body mass by species and sex')
plt.xlabel('Species')
plt.ylabel('Body mass')
plt.legend(title='Sex')
plt.show()
```

```python
sns.kdeplot(data=t['body_mass_g'])
plt.title('body mass Distribution')
plt.xlabel('body mass')
plt.show()
```
